```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# Base URL of Ola Electric
BASE_URL = 'https://www.olaelectric.com/'

def fetch_ola_models():
    try:
        # Send a GET request to the base URL
        response = requests.get(BASE_URL)
        response.raise_for_status()  # Raise an exception for HTTP errors

        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all anchor tags with the specified class
        model_tags = soup.find_all('a', class_='mgz-link mgz-btn')

        # Extract model names and their relative URLs
        models = []
        for tag in model_tags:
            name = tag.get_text(strip=True)
            relative_url = tag.get('href')
            if name and relative_url:
                full_url = urljoin(BASE_URL, relative_url)
                models.append({'name': name, 'url': full_url})

        return models

    except requests.RequestException as e:
        print(f"Error fetching data: {e}")
        return []

# Execute the function and print the results
if __name__ == "__main__":
    ola_models = fetch_ola_models()
    for model in ola_models:
        print(f"{model['name']}: {model['url']}")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Setup Chrome
options = Options()
options.add_argument("--headless")  # Remove this if you want to see browser
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 15)

driver.get("https://book.olaelectric.com/")

# Wait for models to load
wait.until(EC.presence_of_element_located((By.CLASS_NAME, "variants-menu")))
model_buttons = driver.find_elements(By.CSS_SELECTOR, ".variants-menu .variants-menu-item")

bike_specs = {}

def handle_overlay_if_present():
    """Handle overlay if it appears."""
    try:
        overlay = WebDriverWait(driver, 3).until(
            EC.presence_of_element_located((By.CLASS_NAME, "overlay"))
        )
        print("üîí Overlay appeared. Handling it...")

        pin_input = driver.find_element(By.ID, "pin")
        pin_input.send_keys("110001")

        continue_button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, ".next-button:not([disabled])"))
        )
        continue_button.click()

        WebDriverWait(driver, 10).until(
            EC.invisibility_of_element_located((By.CLASS_NAME, "overlay"))
        )
        print("‚úÖ Overlay closed.")

    except:
        pass  # No overlay

for i, btn in enumerate(model_buttons):
    try:
        driver.execute_script("arguments[0].scrollIntoView();", btn)
        btn.click()
        time.sleep(2)

        handle_overlay_if_present()

        wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "model-selection-box-np")))
        variant_boxes = driver.find_elements(By.CLASS_NAME, "model-selection-box-np")

        for box in variant_boxes:
            # Default values
            model_name = battery = certified_range = top_speed = acceleration = "NA"

            try:
                title_section = box.find_element(By.CLASS_NAME, "varient-title-section").text
                title_parts = title_section.split("‚Ä¢")
                model_name = title_parts[0].strip()
                if len(title_parts) > 1:
                    battery = title_parts[1].replace("NEW", "").strip()
            except:
                pass

            try:
                certified_range = box.find_element(By.CLASS_NAME, "range-np").find_element(By.CLASS_NAME, "val-np").text.strip()
            except:
                pass

            speed_blocks = box.find_elements(By.CLASS_NAME, "speed-np")
            try:
                top_speed = speed_blocks[0].find_element(By.CLASS_NAME, "val-np").text.strip()
            except:
                pass

            try:
                acceleration = speed_blocks[1].find_element(By.CLASS_NAME, "val-np").text.strip()
            except:
                pass

            # Save data
            if model_name not in bike_specs:
                bike_specs[model_name] = {}

            bike_specs[model_name][battery] = {
                "Certified Range": certified_range,
                "Top Speed": top_speed,
                "0-40 km/h": acceleration
            }

            print(f"\n‚úÖ {model_name} ‚Äî {battery}")
            print(f"Range: {certified_range}")
            print(f"Top Speed: {top_speed}")
            print(f"0-40 km/h: {acceleration}")

    except Exception as e:
        print(f"‚ùå Error with model {i}: {e}")

driver.quit()

# Output summary
print("\nüìä All Bike Specs:")
for model, variants in bike_specs.items():
    print(f"\nüõµ {model}")
    for battery, specs in variants.items():
        print(f" üîã {battery}")
        for k, v in specs.items():
            print(f"   - {k}: {v}")

# Save to CSV
import pandas as pd
import os
from datetime import datetime
# Create a DataFrame from the bike_specs dictionary
data = []
for model, variants in bike_specs.items():
    for battery, specs in variants.items():
        data.append({
            'Model': model,
            'Battery': battery,
            'Certified Range': specs['Certified Range'],
            'Top Speed': specs['Top Speed'],
            '0-40 km/h': specs['0-40 km/h']
        })
df = pd.DataFrame(data)
# Save DataFrame to CSV
output_dir = "ola_bike_specs"
os.makedirs(output_dir, exist_ok=True)
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
output_file = os.path.join(output_dir, f"ola_Scooty_specs_{timestamp}.csv")
df.to_csv(output_file, index=False)
print(f"\n‚úÖ Data saved to {output_file}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Setup Chrome options
options = Options()
# options.add_argument("--headless")  # Run in headless mode
options.add_argument("--window-size=1920,1080")

# Initialize the WebDriver
driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 15)

try:
    # Step 1: Navigate to the TVS Vehicles page
    driver.get("https://www.tvsmotor.com/our-products/vehicles")

    # Step 2: Click on the 'Electric' filter tab
    electric_tab = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "li.filterVehicle[data-id='#set4']")))
    electric_tab.click()

    # Step 3: Wait for the TVS X image to be clickable and click it
    tvs_x_image = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "img[src*='TVSX-EntryPoints/Our-Product-Page']")))
    tvs_x_image.click()

    # Step 4: Wait for the specifications section to load
    specs_list = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "ul.specsList")))

    # Step 5: Extract specifications
    specs_items = specs_list.find_elements(By.TAG_NAME, "li")
    vehicle_details = {
        "Real Range (Eco)": "NA",
        "Battery Capacity": "NA",
        "Top Speed": "NA"
    }

    for item in specs_items:
        text = item.text.strip()
        if "Real Range" in text:
            vehicle_details["Real Range (Eco)"] = item.find_element(By.TAG_NAME, "strong").text
        elif "Battery Capacity" in text:
            vehicle_details["Battery Capacity"] = item.find_element(By.TAG_NAME, "strong").text
        elif "Top Speed" in text:
            vehicle_details["Top Speed"] = item.find_element(By.TAG_NAME, "strong").text

    # Step 6: Print the extracted details
    print("TVS X Electric Scooter Specifications:")
    for key, value in vehicle_details.items():
        print(f"{key}: {value}")

finally:
    # Close the browser
    driver.quit()
```
